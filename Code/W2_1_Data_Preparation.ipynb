### 1.3 Categorize each entry by rating group

Rule:
- negative: 1.0 to 2.5 (inclusive)
- neutral: exactly 3.0
- positive: 3.5 to 5.0 (inclusive)

```python
def rating_group(x: float) -> str:
  if pd.isna(x):
    return "missing"
  if 1.0 <= x <= 2.5:
    return "negative"
  if x == 3.0:
    return "neutral"
  if 3.5 <= x <= 5.0:
    return "positive"
  return "other"
```

```python
df["rating_group"] = df["student_star"].apply(rating_group)
```

```python
df
```

### 1.4 Parse student tags into unique tag labels

In RMP, stu_tags are often phrases separated by TWO OR MORE spaces, not single spaces.

Example: "TEST HEAVY  LECTURE HEAVY  SKIP CLASS? YOU WON'T PASS."

Here each tag is a multi-word phrase, and tags are separated by "  " (double spaces).

We split on 2+ whitespace characters to keep multi-word tags intact.


```python
import re

def split_tags(tag_str: str) -> list:
  if pd.isna(tag_str):
    return []
  s = str(tag_str).strip()
  if not s:
    return []
  # split tags by two or more whitespace characters (robust to variable spacing)
  parts = re.split(r"\s{2,}", s)
  # clean each tag phrase: strip and drop empties
  parts = [p.strip() for p in parts if p and p.strip()]
  return parts

df["tag_list"] = df["stu_tags"].apply(split_tags)
```

```python
# Explode tag_list to one tag per row (for counting and uniques).
tags_long = df[["rating_group", "tag_list"]].explode("tag_list").rename(columns={"tag_list": "tag_label"})
tags_long["tag_label"] = tags_long["tag_label"].fillna("").astype(str).str.strip()
tags_long = tags_long[tags_long["tag_label"] != ""]  # remove empty tags
```

### 1.5 Summary tables


#### 1.5.1 Overall uniques

```python
overall_uniques = pd.DataFrame([{
   "n_rows": len(df),
    "unique_professors": df["professor_name"].nunique(dropna=True),
    "unique_schools": df["school_name"].nunique(dropna=True),
    "unique_departments": df["department_name_clean"].nunique(dropna=True),
    "unique_tag_labels": tags_long["tag_label"].nunique(dropna=True),
    "stu_star_nonmissing": df["student_star"].notna().sum(),
    "stu_star_mean": df["student_star"].mean(),
    "stu_star_median": df["student_star"].median(),

}])
```

##### 1.5.2 Group-level summary by rating group

```python
group_uniques = (
    df.groupby("rating_group", dropna=False)
      .agg(
          n_rows=("student_star", "size"),
          unique_professors=("professor_name", "nunique"),
          unique_schools=("school_name", "nunique"),
          unique_departments=("department_name_clean", "nunique"),
          stu_star_nonmissing=("student_star", lambda x: x.notna().sum()),
          stu_star_mean=("student_star", "mean"),
          stu_star_median=("student_star", "median"),
          stu_star_min=("student_star", "min"),
          stu_star_max=("student_star", "max"),
      )
      .reset_index()
)
```

```python
# Add "unique_tag_labels" per group using the exploded tags table.
tag_uniques_by_group = (
    tags_long.groupby("rating_group")["tag_label"]
            .nunique()
            .reset_index()
            .rename(columns={"tag_label": "unique_tag_labels"})
)

group_uniques = group_uniques.merge(tag_uniques_by_group, on="rating_group", how="left")
group_uniques["unique_tag_labels"] = group_uniques["unique_tag_labels"].fillna(0).astype(int)
```

**Descriptive Analysis of Rating Groups**

We categorized each student comment into three rating groups based on the `student_star` score: **negative** (1.0-2.5), **neutral** (3.0), and **positive** (3.5-5.0). Out of 19,685 total comments, the vast majority (13,615, \~69\%) fall into the positive group, reflecting generally high student satisfaction. The negative group comprises 4,862 comments (\~25\%), while the neutral group is the smallest with 1,195 comments (\~6\%).

Comments in the positive group are associated with 2,481 unique professors across 1,064 schools and 251 unique tag labels, and exhibit a mean `student_star` of 4.62 and median of 5.0. The negative group includes 1,544 professors from 807 schools, with a mean rating of 1.56 and median of 1.5, indicating strong dissatisfaction. The neutral group, with mean and median exactly at 3.0, spans 802 professors and 490 schools.

This distribution indicates a pronounced skew toward positive evaluations, but substantial variability exists across professors and schools. Notably, the negative and neutral groups still cover a wide range of departments and tag labels, suggesting that even lower-rated comments are spread across diverse academic contexts rather than being isolated to a few instructors or courses. This preliminary categorization will serve as the basis for further analysis of text features and gendered perception in student evaluations.
